```python
import pandas as pd
import numpy as np
from pandas.api.types import CategoricalDtype
from plotnine import *
from plotnine.data import mpg,diamonds
%matplotlib inline    
```

```python
df = pd.read_csv("amazon.csv")
```

```python
df.head()
```

```python
(ggplot(df) + aes(x='Class') + geom_bar(size=5) + coord_flip() + labs(title='Amazon Dataset') )                                 
```

```python
(ggplot(df.head(50)) + aes(x='Reviews', y='Author', color='Class') + geom_point() 
 + labs(title='Author Vs Reviews', x='Reviews', y='Author'))
```

```python
ggplot(data = df, mapping = aes(x = 'Class', y = 'Reviews')) + geom_boxplot()
```

```python
ggplot(df.head(50)) + geom_point(mapping = aes(x ='User Rating', y ='Author'))
```

```python
ggplot(df.head(50)) + geom_point(aes(x='Year', y='Name', color='Class'))+scale_color_brewer(type='qual')
```

```python
ggplot(df) + aes(x='User Rating')+ stat_bin(bins=10) + geom_bar()
```

```python
ggplot(df.head(50), aes(x='Year', y='Author', z="Reviews", fill="Class", shape='Class')) + geom_point()
```

```python
# facet plot with labeller
(
    ggplot(df.head(50), aes(x='Reviews', y='Author'))
    + geom_point()
    + facet_wrap('Class', labeller = 'label_both')
    + labs(x='Reviews', y='Author')
)
```

```python
# facet plot with two variables on one facet
(
    ggplot(df.head(50), aes(x='Year', y='Price'))
    + geom_point()
    + facet_wrap('~ Class + Reviews') # use ~ + to add additional faceting variables
    + labs(x='Year', y='Price')
)
```

```python
# facet plot with two variables on one facet
(
    ggplot(df.head(50), aes(x='Price'))
    + geom_histogram()
    + facet_wrap('User Rating') # use ~ + to add additional faceting variables
)
```

```python
ggplot(df,aes(x='User Rating', color='Class')) + geom_density()
```

```python
ggplot(df, aes(x='Year', y='Reviews', color='User Rating')) + geom_jitter() 
```
