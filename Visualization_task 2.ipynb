```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme(style="ticks", color_codes=True)
```

```python
df=pd.read_csv("iris_new.csv")
df
```

```python
sns.histplot(data=df, x='petal.length', hue='class', multiple='stack')
```

```python
sns.displot(data=df, x='sepal.width', hue='class', col='class')
```

```python
sns.pairplot(df, hue="class")
```

```python
sns.boxplot(x="sepal.width", hue="class", data=df)
sns.despine(offset=10, trim=True)
```

```python
g = sns.PairGrid(df)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(df, hue="class")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
```

```python
import matplotlib.pyplot as plt
# CGPA data
cgpa = [2.5, 3.7, 3.0, 3.5]
semesters = ['First', 'Second', 'Third', 'Fourth']
# Create the pie chart
plt.pie(cgpa, labels=semesters, autopct='%1.1f%%')
# Set aspect ratio to make the pie chart a circle
plt.axis('equal')
# Add a title
plt.title('CGPA Distribution by Semester')
# Display the pie chart
plt.show()
```
