```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x=[1,2,3]
y=[1,2,3]
plt.plot(x,y)
plt.title(" First Plot")
plt.xlabel(" X Frequency ")
plt.ylabel(" Y Frequency ")
plt.legend(["Y"])
```

```python
x=[1,2,3]
y=[1,4,9]
z=[8,5,0]
plt.plot(x,y,z)
plt.title("First Plot")
plt.xlabel(" X Frequency ")
plt.ylabel(" Y Frequency ")
plt.legend([" Y ", " Z "])
plt.show()
```

```python
x=["Sana","Aniyat","Mohsin"]
y=[20,30.5,18]
```

```python
plt.bar(x,y)
plt.title("3D")
plt.xlabel("Names")
plt.ylabel("Age")
plt.legend(["x","y"])
```
